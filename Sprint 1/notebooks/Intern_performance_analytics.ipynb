```python
import pandas as pd
# Loading dataset
df = pd.read_csv("Interns__data.csv")
```

```python
print(df.info()) # Basic info
```

```python
print(df.describe()) # Basic info
```

```python
# check for Missing values
print(df.isnull().sum())
```

```python
# Filling missing values
df['attendance'] = df['attendance'].fillna(df['attendance'].mean())
df['mentor_feedback'] = df['mentor_feedback'].fillna(df['mentor_feedback'].median())
```

```python
# Removing the duplicates and invalid attendance
df = df.drop_duplicates()
df = df[df['attendance'] <= 100]
```

```python
# Cleaning the text columns
df['Internship domain'] = df['Internship domain'].str.strip()
df['sprint_status'] = df['sprint_status'].str.strip().str.title()
```

```python
#Validating the sprint number
df = df[(df['sprint_number'] >= 1) & (df['sprint_number'] <= 4)]
```

```python
# Removing invalid attendance
df = df[df['attendance'] <= 100]
```

```python
df["total_sprints_completed"] = df.apply(
    lambda row: row["sprint_number"] if row["sprint_status"] == "Completed"
    else row["sprint_number"] - 1,
    axis=1
)
```

```python
# Creating new columns
df["task_completion_rate"] = (df["tasks_completed"] / df["tasks_assigned"]) * 100
df["efficiency_score"] = (df["performance_score"] * 0.6) + (df["attendance"] * 0.4)
```

```python
#categorizing the performance of the interns
def category(score):
    if score >= 8:
        return "High"
    elif score >= 5:
        return "Medium"
    else:
        return "Low"

df["performance_category"] = df["performance_score"].apply(category)
```

```python
df.to_csv("cleaned_Interns_data.csv", index=False)  # Final cleaned dataset
```
